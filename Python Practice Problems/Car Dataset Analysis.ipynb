```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(r'cars_2010_2020.csv')
```

```python
#df1.head(5)
```

```python
df1['Price (INR)'] = (df1['Price (USD)'] * 88.77).round(0)
df1.head(5)
```

```python
piv_table = df1.pivot_table(
    values= 'Engine Size (L)',
    index= 'Make',
    columns= 'Price (INR)',
    aggfunc= 'mean'
)
piv_table = piv_table.fillna('O')
print(piv_table)
```

Basic Analytical Questions

```python
# Count of all cars in the data
total_cars = df1['Model'].nunique()
total_cars
```

```python
# Count of all Manufacturers in the dataset
all_makers = df1['Make'].nunique()
all_makers
```

```python
# Count of top 5 cars in the data
counts = df1['Model'].value_counts().head(5)
counts
```

```python
# Average Price of Fuel
avg_price_of_fuel = df1.groupby('Fuel Type')['Price (INR)'].mean().round(2)
avg_price_of_fuel
```

```python
# Minimum Engine Size
min_eng_size = df1['Engine Size (L)'].min()
min_eng_size
```

```python
# Maximum Engine Size
max_eng_size = df1['Engine Size (L)'].max()
max_eng_size
```

Advance Analytical Questions

```python
# Average Price Per Car Maker
avg_price = df1.groupby('Make')['Price (INR)'].mean().round(2)
avg_price
```

```python
# Find the total number of cars produced each year
count_of_cars = df1.groupby('Year').size().to_frame('Count of Cars')
count_of_cars
```

```python
# Show the count of cars grouped by fuel type and Maker
car_by_fuel_and_maker = df1.groupby(['Fuel Type', 'Make']).size().to_frame('Count of Cars')
car_by_fuel_and_maker
```

```python
# Which 5 cars have the highest price in INR?
cars_w_highest_price = df1.sort_values(by='Price (INR)', ascending=False)
cars_w_highest_price
```

```python
# Find cars where engine size is greater than 3.0 liters
large_eng = df1[df1['Engine Size (L)'] > 3].head(5)
large_eng
```

Pivot/ Aggregation

```python
# Create a pivot summary of mean price (USD) by Make and Fuel Type.
pt = df1.pivot_table(
    index= 'Make',
    columns= 'Fuel Type',
    values= 'Price (USD)',
    aggfunc='mean'
).round(2)
pt
```

```python
# Create a pivot table showing count of cars per Manufacturer and per Year.
pt2 = df1.pivot_table(
    index= 'Make',
    columns= 'Year',
    values= 'Price (INR)',
    aggfunc='count'
)
pt2
```

Conditional / Filtering

```python
# Label cars as ‘Premium’ if Price (USD) > 50,000, else ‘Standard’.
df1['Segment'] = np.where(df1['Price (USD)'] > 50000, 'Premium', 'Standard')
df1.head(5)
```

```python
# Filter all Electric cars costing more than ₹30,00,000 INR.
elc_cars = df1[(df1['Fuel Type'] == 'Electric') & (df1['Price (INR)'] > 3000000)]
elc_cars.head(10)
```

Ranking

```python
# Rank cars within each Maker based on Price (USD).
df1['car_maker_rank'] = df1.groupby('Make')['Price (USD)'].rank(method= 'dense', ascending= False)
df1.head(10)
```
